
## Find Words Containing Character

### Solution 

```python
class Solution(object):
    def findWordsContaining(self, words, x):
        ans = []
        a = len(words)
        for i in range(a):
            if x in words[i]:
                ans += [i]
        return ans                
        """
        :type words: List[str]
        :type x: str
        :rtype: List[int]
        """
        
```

```python
words = ["leet","code","m"]
x = "e"
a = len(words)
ans = []
for i in range(a):
    if x in words[i]:
        ans += [i]
print(ans)
```
