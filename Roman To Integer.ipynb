for coverting roman to integer the solution i come with is using if else condition there are multiple type  
we can solve this problem 

```python
class Solution(object):
    def romanToInt(self, s):
        ans = 0
        n = len(s)
        for i in range(n):
            if s[i] == "I":
                value = 1
            elif s[i] == "V":
                value = 5
            elif s[i] == "X":
                value = 10
            elif s[i] == "L":
                value = 50
            elif s[i] == "C":
                value = 100
            elif s[i] == "D":
                value = 500
            elif s[i] == "M":
                value = 1000

            if i+ 1 < n:
                if value < (
                    1 if s[i+1] =="I" else
                    5 if s[i+1] =="V" else
                    10 if s[i+1] =="X" else
                    50 if s[i+1] =="L" else
                    100 if s[i+1] =="C" else
                    500 if s[i+1] =="D" else
                    1000
                ):
                    ans -= value
                    continue
            ans += value
        return ans

r = Solution()
print(r.romanToInt("XIIIV"))
```
